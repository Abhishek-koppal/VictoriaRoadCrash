```sql
SELECT
  location,
  fatal_count,
  total_count,
  fatality_rate
FROM (
  SELECT
    LGA_NAME AS location,
    SUM(FATALITY) AS fatal_count,
    COUNT(*) AS total_count,
    (SUM(FATALITY) / COUNT(*)) AS fatality_rate,
    RANK() OVER (ORDER BY (SUM(FATALITY) / COUNT(*)) DESC) AS rnk
  FROM roadcrashvic
  GROUP BY LGA_NAME
)
WHERE rnk BETWEEN 2 AND 11
ORDER BY fatality_rate DESC
```
